```python
from langgraph.store.memory import InMemoryStore
import uuid
from langchain_huggingface import HuggingFaceEmbeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

```python
key1 = str(uuid.uuid4())
key2 = str(uuid.uuid4())
key3 = str(uuid.uuid4())
key4 = str(uuid.uuid4())
```

```python
store = InMemoryStore()
namespace1 = ("user","u1") # created in the form of a tuple
namespace2 = ("user","u2") # created in the form of a tuple
```

# Creating a memory

```python
store.put(namespace1, key1, {"data": "User like eating burgers."})
store.put(namespace1, key2, {"data": "User likes high contrast dark mode images."})
store.put(namespace2, key3, {"data": "User like eating dosas."})
store.put(namespace2, key4, {"data": "User likes light mode."})
```

# Retrieval

```python
store.get(namespace1, key2).value['data']
```

```python
store.get(namespace2, key4).value['data']
```

# Retrieving all memories

```python
items = store.search(namespace2)

for item in items:
    print(item.value['data'])
```

# Performing Semantic Search

```python
storer = InMemoryStore(index={'embed': embeddings, 'dims':1536})
```

```python
namespace3 = ('users', 'u3')
```

```python
storer.put(namespace3, "1", {"data": "User prefers concise answers over long explanations"})
storer.put(namespace3, "2", {"data": "User likes examples in Python"})
storer.put(namespace3, "3", {"data": "User usually works late at night"})
storer.put(namespace3, "4", {"data": "User prefers dark mode in applications"})
storer.put(namespace3, "5", {"data": "User is learning machine learning"})
storer.put(namespace3, "6", {"data": "User dislikes overly theoretical explanations"})
storer.put(namespace3, "7", {"data": "User prefers step-by-step reasoning"})
storer.put(namespace3, "8", {"data": "User is based in India"})
storer.put(namespace3, "9", {"data": "User likes real-world analogies"})
storer.put(namespace3, "10", {"data": "User prefers bullet points over paragraphs"})
```

```python
items = storer.search(namespace3, query="what is the user currently learning", limit=1)

for item in items:
    print(item.value['data'])
```
